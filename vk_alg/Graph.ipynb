```python
import ParsFromVK
import pymorphy2
import string
import pickle

def check_Gramme(word, morth=pymorphy2.MorphAnalyzer()): # Граммема слова
    
    return morth.parse(word)[0].tag.POS


def get_normal_form(word, morth=pymorphy2.MorphAnalyzer()): # Нормальная форма слова
    
    return morth.parse(word)[0].normal_form


def delete_punctuation(word): # Удаление всех знаков пунктуации
    
    tt = str.maketrans(dict.fromkeys(string.punctuation))
    
    return word.translate(tt)


forbidden_grammes = {'NPRO', 'PRCL', 'INTJ', 'CONJ', 'PREP', 'PRED', 'PNCT'}  # Запрещенные граммемы
forbidden_words = {'title', 'True', 'False', 'id', 'false', 'true'} # Запрещенные слова
professions = {'Агроном', 'Бухгалтер', 'Грузчик', 'Менеджер', 'Облицовщик', 'Пожарный', 'Учитель'} # Профессии

#print(get_normal_form('18121990'), check_Gramme('18121990'))

number_vertex = {} # Номер вершины по названию
vertex_edges = {} # Соседи вершины

num = 1
sumcnt = 0
count_occ = {}

# Агрономы

name = 'Агроном'
number_vertex[name] = num
vertex_edges[name] = []
num += 1
cnt = {} # число раз которое слово встретилось
file = open('Агрономы.txt', 'r')
for line in file:
    cur_id = line
    if ParsFromVK.is_close(cur_id) == True:
        continue
    sumcnt += 1
    main_information, groups = ParsFromVK.pars_from_vk(cur_id)
    for group in groups:
        words = group.split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None and word not in professions:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                    count_occ[word] = set()
                    count_occ[word].add(cur_id)
                else:
                    cnt[word] += 1
                    count_occ[word].add(cur_id)
    for key in main_information:
        words = str(main_information[key]).split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None and word not in professions:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                    count_occ[word] = set()
                    count_occ[word].add(cur_id)
                else:
                    cnt[word] += 1
                    count_occ[word].add(cur_id)
temp = []
for key in cnt:
    if len(key) != 0 and key not in forbidden_words:
        temp.append((key, cnt[key]))
temp.sort(key=lambda point: (-point[1]))
for i in range(0, 130):
    curName = temp[i][0]
    curCnt = temp[i][1]
    if curName not in number_vertex:
        number_vertex[curName] = num
        num += 1
        vertex_edges[curName] = []
    vertex_edges[curName].append((name, curCnt))
    vertex_edges[name].append((curName, curCnt))

print('Agronom finish')
    
# Бухгалтер

name = 'Бухгалтер'
number_vertex[name] = num
vertex_edges[name] = []
num += 1
cnt = {} # число раз которое слово встретилось
file = open('Бухгалтера.txt', 'r')
for line in file:
    cur_id = line
    if ParsFromVK.is_close(cur_id) == True:
        continue
    sumcnt += 1
    main_information, groups = ParsFromVK.pars_from_vk(cur_id)
    for group in groups:
        words = group.split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None and word not in professions:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                    count_occ[word] = set()
                    count_occ[word].add(cur_id)
                else:
                    cnt[word] += 1
                    count_occ[word].add(cur_id)
    for key in main_information:
        words = str(main_information[key]).split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None and word not in professions:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                    count_occ[word] = set()
                    count_occ[word].add(cur_id)
                else:
                    cnt[word] += 1
                    count_occ[word].add(cur_id)
temp = []
for key in cnt:
    if len(key) != 0 and key not in forbidden_words:
        temp.append((key, cnt[key]))
temp.sort(key=lambda point: (-point[1]))
for i in range(0, 130):
    curName = temp[i][0]
    curCnt = temp[i][1]
    if curName not in number_vertex:
        number_vertex[curName] = num
        num += 1
        vertex_edges[curName] = []
    vertex_edges[curName].append((name, curCnt))
    vertex_edges[name].append((curName, curCnt))

print('Accountant finish')

# Грузчик

name = 'Грузчик'
number_vertex[name] = num
vertex_edges[name] = []
num += 1
cnt = {} # число раз которое слово встретилось
file = open('Грузчики.txt', 'r')
for line in file:
    cur_id = line
    #print(cur_id)
    if ParsFromVK.is_close(cur_id) == True:
        continue
    sumcnt += 1
    main_information, groups = ParsFromVK.pars_from_vk(cur_id)
    for group in groups:
        words = group.split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None and word not in professions:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                    count_occ[word] = set()
                    count_occ[word].add(cur_id)
                else:
                    cnt[word] += 1
                    count_occ[word].add(cur_id)
    for key in main_information:
        words = str(main_information[key]).split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None and word not in professions:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                    count_occ[word] = set()
                    count_occ[word].add(cur_id)
                else:
                    cnt[word] += 1
                    count_occ[word].add(cur_id)
temp = []
for key in cnt:
    if len(key) != 0 and key not in forbidden_words:
        temp.append((key, cnt[key]))
temp.sort(key=lambda point: (-point[1]))
for i in range(0, 130):
    curName = temp[i][0]
    curCnt = temp[i][1]
    if curName not in number_vertex:
        number_vertex[curName] = num
        num += 1
        vertex_edges[curName] = []
    vertex_edges[curName].append((name, curCnt))
    vertex_edges[name].append((curName, curCnt))

print('Loader finish')

# Менеджер

name = 'Менеджер'
number_vertex[name] = num
vertex_edges[name] = []
num += 1
cnt = {} # число раз которое слово встретилось
file = open('Менеджер.txt', 'r')
for line in file:
    cur_id = line
    if ParsFromVK.is_close(cur_id) == True:
        continue
    sumcnt += 1
    main_information, groups = ParsFromVK.pars_from_vk(cur_id)
    for group in groups:
        words = group.split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None and word not in professions:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                    count_occ[word] = set()
                    count_occ[word].add(cur_id)
                else:
                    cnt[word] += 1
                    count_occ[word].add(cur_id)
    for key in main_information:
        words = str(main_information[key]).split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None and word not in professions:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                    count_occ[word] = set()
                    count_occ[word].add(cur_id)
                else:
                    cnt[word] += 1
                    count_occ[word].add(cur_id)
temp = []
for key in cnt:
    if len(key) != 0 and key not in forbidden_words:
        temp.append((key, cnt[key]))
temp.sort(key=lambda point: (-point[1]))
for i in range(0, 130):
    curName = temp[i][0]
    curCnt = temp[i][1]
    if curName not in number_vertex:
        number_vertex[curName] = num
        num += 1
        vertex_edges[curName] = []
    vertex_edges[curName].append((name, curCnt))
    vertex_edges[name].append((curName, curCnt))

print('Manager finish')

# Облицовщик

name = 'Облицовщик'
number_vertex[name] = num
vertex_edges[name] = []
num += 1
cnt = {} # число раз которое слово встретилось
file = open('Облицовщик.txt', 'r')
for line in file:
    cur_id = line
    if ParsFromVK.is_close(cur_id) == True:
        continue
    sumcnt += 1
    main_information, groups = ParsFromVK.pars_from_vk(cur_id)
    for group in groups:
        words = group.split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None and word not in professions:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                    count_occ[word] = set()
                    count_occ[word].add(cur_id)
                else:
                    cnt[word] += 1
                    count_occ[word].add(cur_id)
    for key in main_information:
        words = str(main_information[key]).split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None and word not in professions:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                    count_occ[word] = set()
                    count_occ[word].add(cur_id)
                else:
                    cnt[word] += 1
                    count_occ[word].add(cur_id)
temp = []
for key in cnt:
    if len(key) != 0 and key not in forbidden_words:
        temp.append((key, cnt[key]))
temp.sort(key=lambda point: (-point[1]))
for i in range(0, 130):
    curName = temp[i][0]
    curCnt = temp[i][1]
    if curName not in number_vertex:
        number_vertex[curName] = num
        num += 1
        vertex_edges[curName] = []
    vertex_edges[curName].append((name, curCnt))
    vertex_edges[name].append((curName, curCnt))

print('Facing finish')

# Пожарный

name = 'Пожарный'
number_vertex[name] = num
vertex_edges[name] = []
num += 1
cnt = {} # число раз которое слово встретилось
file = open('Пожарные.txt', 'r')
for line in file:
    cur_id = line
    if ParsFromVK.is_close(cur_id) == True:
        continue
    sumcnt += 1
    main_information, groups = ParsFromVK.pars_from_vk(cur_id)
    for group in groups:
        words = group.split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None and word not in professions:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                    count_occ[word] = set()
                    count_occ[word].add(cur_id)
                else:
                    cnt[word] += 1
                    count_occ[word].add(cur_id)
    for key in main_information:
        words = str(main_information[key]).split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None and word not in professions:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                    count_occ[word] = set()
                    count_occ[word].add(cur_id)
                else:
                    cnt[word] += 1
                    count_occ[word].add(cur_id)
temp = []
for key in cnt:
    if len(key) != 0 and key not in forbidden_words:
        temp.append((key, cnt[key]))
temp.sort(key=lambda point: (-point[1]))
for i in range(0, 130):
    curName = temp[i][0]
    curCnt = temp[i][1]
    if curName not in number_vertex:
        number_vertex[curName] = num
        num += 1
        vertex_edges[curName] = []
    vertex_edges[curName].append((name, curCnt))
    vertex_edges[name].append((curName, curCnt))

print('Firemen finish')

# Учитель

name = 'Учитель'
number_vertex[name] = num
vertex_edges[name] = []
num += 1
cnt = {} # число раз которое слово встретилось
file = open('Учителя.txt', 'r')
for line in file:
    cur_id = line
    if ParsFromVK.is_close(cur_id) == True:
        continue
    sumcnt += 1
    main_information, groups = ParsFromVK.pars_from_vk(cur_id)
    for group in groups:
        words = group.split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None and word not in professions:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                    count_occ[word] = set()
                    count_occ[word].add(cur_id)
                else:
                    cnt[word] += 1
                    count_occ[word].add(cur_id)
    for key in main_information:
        words = str(main_information[key]).split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None and word not in professions:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                    count_occ[word] = set()
                    count_occ[word].add(cur_id)
                else:
                    cnt[word] += 1
                    count_occ[word].add(cur_id)
temp = []
for key in cnt:
    if len(key) != 0 and key not in forbidden_words:
        temp.append((key, cnt[key]))
temp.sort(key=lambda point: (-point[1]))
for i in range(0, 130):
    curName = temp[i][0]
    curCnt = temp[i][1]
    if curName not in number_vertex:
        number_vertex[curName] = num
        num += 1
        vertex_edges[curName] = []
    vertex_edges[curName].append((name, curCnt))
    vertex_edges[name].append((curName, curCnt))

print('Teacher finish')

file_1 = open('data1', 'wb')
res = []
res.append(vertex_edges)
res.append(count_occ)
res.append(sumcnt)
pickle.dump(res, file_1)
```
